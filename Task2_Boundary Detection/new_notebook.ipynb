```python
img="./images/apple.jpg"
image=cv2.imread(img)
image= cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
```

```python
#parameters
alpha=30
beta=20
gamma=7
contour_points=60
```

```python
snake=contour(image,alpha,beta,gamma,contour_points)
for j in range(100):
    snake.fit_snake()
snake.draw_contour()
```

```python
len=snake.contour_len()
print("perimeter:","",len)
area=snake.contour_area()
print("area:","",area)
```
